```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("House Price Prediction Dataset.csv")
df.info()
```

```python
summary_statistics = df[["Price", "Area"]].describe()
```

```python
df["prize_per_sqft"] = df["Price"] / df["Area"]
```

```python
plt.scatter(x= df["Price"], y= df["Area"], alpha= 0.5, color= "blue")
plt.xlabel("Price")
plt.ylabel("Square Footage")
plt.title("Price VS Square footage.")
plt.show()
```
