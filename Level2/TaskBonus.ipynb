```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"A:\Your level\Level2\tested.csv")
df.drop(columns= ["Name", "Sex", "Ticket", "Cabin", "Embarked"], axis=1,inplace=True)
```

```python
correlation_matrix = df.corr()
heatmap = sns.heatmap(correlation_matrix, annot = True, cmap ="magma", fmt = ".2f", linewidths= 1)
plt.show(heatmap)
```
