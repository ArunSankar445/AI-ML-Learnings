```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
data = np.random.binomial(n= 10, p= 0.5, size= 1000)
```

```python
plt.hist(data, bins= range(12), density= True)
plt.xlabel("Number of Successes")
plt.ylabel("Probability Density")
plt.title("Binomial Distribution (n=10, p=0.5, 1000 experiments)")
plt.show()
```

```python
mean = np.mean(data)
print(f"Simulated mean: {mean}")
print("Matches binomial expectation (n x p = 5).")
```
